# Get the dbbact ontology terms

```python
import requests
```

```python
db='http://127.0.0.1:5001'
```

```python
res=requests.get(db+'/ontology/get_all_terms',json={'ontologyid':1, 'min_term_id':0})
```

```python
res=res.json()
```

```python
len(res)
```

```python
print(max(res, key=res.get))
```

```python
res['streptomycin']
```

## save to a pickle file

```python
max_id = res[max(res, key=res.get)]
```

```python
res['_dbbact_calour_max_id'] = max_id
```

```python
import pickle
```

```python
with open('test.pickle','wb') as ofl:
    pickle.dump(res,ofl)
```
